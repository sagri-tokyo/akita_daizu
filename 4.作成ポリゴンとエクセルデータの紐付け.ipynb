496のポリゴンに現地調査結果120箇所の情報を追加したポリゴンは、496の内半数近くはアグリプロから提供されたエクセルデータと突合できていない＝作付け年のデータが入っていないので、もう一度エクセルと496ポリゴンの突合をする


# 求めていた情報が追加されたGEOJSONはこの処理を通して生成済み　このコードはもう動かす必要はない

この処理をする前のポリゴンは念の為'/workspaces/akita_daizu/data/bu 474+120圃場 copy.geojson'で保管

```python
import geopandas as gpd
gpd_data = gpd.read_file('/workspaces/akita_daizu/data/474+120圃場.geojson')
```

```python
import pandas as pd
agri_excel = pd.read_csv('/workspaces/akita_daizu/data/8月25日　アグリ　面積 R3.csv')
```

```python
gpd_data
```

```python
agri_excel
```

```python
#496ポリゴンと結合するため、エクセルの地名地番＿番号列に件名と大字名を付与
agri_excel["Address"] = "秋田県八峰町"+ agri_excel["地名地番_漢字"]
```

```python
agri_excel['Address'] = agri_excel['Address'].str.replace('杉の上', '杉ノ上')#41箇所追加突合できた
agri_excel['Address'] = agri_excel['Address'].str.replace('治郎兵ェ屋敷', '治郎兵エ屋敷')#9箇所追加突合できた
agri_excel['Address'] = agri_excel['Address'].str.replace('形ヶ沢', '形ケ沢')#1箇所追加突合できた
agri_excel['Address'] = agri_excel['Address'].str.replace('沼の尻', '沼ノ尻')#1箇所追加突合できた
agri_excel['Address'] = agri_excel['Address'].str.replace('上ノ野', '上野')#2箇所追加突合できた
```

```python
# とりあえずマージ inner
merged_pd = pd.merge(gpd_data, agri_excel, on=['Address'], how='inner')
```

```python
# 全部の列が必要なわけではないから　いらない列を消してみやすくする
merged_pd.columns
```

```python
# これらのAddresに今まで情報としてくっつけられていなかった作付け年と播種日が含まれる　
# 再度これらの播種日を接続してAddres
merged_pd[['Address','作付年_y', 'H23以前_y','播種月日']]
```

```python
new_merged_pd = pd.merge(gpd_data, merged_pd[['Address','作付年_y', 'H23以前_y','播種月日']], on=['Address'], how='outer')
```

```python
# これで新しい列　'作付年_y', 'H23以前_y','播種月日'の方が更新された列なので、古い'作付年', 'H23以前','播種月?'はDrop
new_merged_pd.columns 
```

```python
drop_columns = ['作付年', 'H23以前','播種月?']
droped_new_merged_pd = new_merged_pd.drop(drop_columns, axis = 1)
```

```python
droped_new_merged_pd
```

```python
droped_new_merged_pd.to_file('/workspaces/akita_daizu/data/474+120圃場.geojson', driver="GeoJSON")  
```
